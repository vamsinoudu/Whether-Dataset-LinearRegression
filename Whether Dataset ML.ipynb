# problem statement -Given Max Temperature as an independent feature, we have to predict min temperature(target feature or dependent variable)

```python
x=df['MaxTemp'].values.reshape(-1,1)
y=df['MinTemp'].values.reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr_pred=lr.predict(x_test)
```

```python
lr_pred
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,lr_pred)*100
```

```python
#Accuracy calculation of test data:
lr.score(x_test,y_test)*100
```

```python
#Checking Accuracy calculation of train data:
lr.score(x_train,y_train)*100
```
